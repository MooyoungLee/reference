### SVM understanding:
- It utilize hiperplanes to separate groups by maximizing margines.
- Kernel trick can be utilized to separate data that is not separable linearly.  
- Kernel trick add features such as squared values or absolute values of feature data.
- As regularization value, 'C', increases the decision boundary become more complex and lead to overfitting.

## Iris dataset

```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris['DESCR'])
```

```python
import pandas as pd
import numpy as np

X = pd.DataFrame(iris.data, columns = iris.feature_names)
y = iris.target.astype(np.int)
X.head()
```

```python
X.isnull().sum()
```

# Scale data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.min(), X_scaled.max()
```

# Support Vector Machine

Accuracy check w/ stratified 10-fold cross-validation

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
clf = SVC()
accuracy = cross_val_score(clf, X_scaled, y, cv = 10)
print('Accuracy: ', np.round(accuracy,3))
print('Mean Accuracy: ', np.mean(accuracy))
```

# Tuning w/ kernel and regularization strength

The smaller value of 'C' specify stronger regularization.

```python
from sklearn.model_selection import RandomizedSearchCV

tuned_parameters = {'kernel' : ['linear','poly','rbf','sigmoid'],
                   'C' : [.0001, .001, .01, .1, .5, 1, 10, 100]}

clf = RandomizedSearchCV(SVC(), tuned_parameters, cv=10, n_iter = 14)
clf.fit(X,y)


## Print Results
print("Best parameters set found on development set:\n")
print(clf.best_params_)
print("Grid scores on development set:\n")
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
          % (mean, std * 2, params))
```

# SVM trained w/ all dataset

```python
clf = SVC(kernel = 'linear', C = 0.5)
clf.fit(X,y)
```

# Support Vector Parameters

```python
X.shape
```

```python
clf.support_vectors_[0:5]
```

```python
print(clf.support_vectors_.shape)
print(clf.n_support_)
```

```python
print(clf.coef_)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
weights = pd.Series(clf.coef_[0], index = X.columns)
weights.plot(kind = 'bar')
```

# Instances chosen by SVM

```python
df_base = X
df_base['target'] = y

df_svm = X.loc[clf.support_,:]
df_svm['target'] = y[clf.support_]
df_svm.head()
```

# Density Plot

<span style = "font-size:1.4em"> The instances chose by support vectors overlaps more since those instances are either from the edge of the class boundaries or from the instances that failed to correctly classify.  

```python
# groupby
df_svm_grp = df_svm.groupby('target')
df_base_grp = df_base.groupby('target')

# kde plot
vars_to_plot = list(X)[0:4]

for v in vars_to_plot:
    plt.figure(figsize = (10,4))
    
    # plot SVM instance
    plt.subplot(1,2,1)
    ax = df_svm_grp[v].plot.kde()
    plt.legend()
    plt.title(v+'_SVM')
    
    # plot original data
    plt.subplot(1,2,2)
    ax = df_base_grp[v].plot.kde()
    plt.legend()
    plt.title(v+'_Original')
```
